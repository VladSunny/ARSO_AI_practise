```python
import random
import numpy as np

def experiment():
    clients = []
    clients.append(9 * 60 + random.expovariate(1/20))

    while clients[-1] < 16 * 60:
        clients.append(clients[-1] + random.expovariate(1/20))
    
    work_time = np.random.uniform(15, 46, len(clients))

    info = {
        'clients': clients,
        'work_time': work_time
    }

    return info

def closing_time(outcome):
    cur_time = 0

    barbers_wait = np.array([0, 0, 0], dtype=float)

    for i in range(len(outcome['clients'])):
        if cur_time >= 16 * 60:
            break
        
        if cur_time < outcome['clients'][i]:
            if outcome['clients'][i] >= 16 * 60:
                break
            barbers_wait -= (outcome['clients'][i] - cur_time)
            barbers_wait[barbers_wait < 0] = 0
            cur_time = outcome['clients'][i]
        
        barbers_wait = np.sort(barbers_wait)
        cur_time += barbers_wait[0]
        barbers_wait -= barbers_wait[0]
        barbers_wait[0] += outcome['work_time'][i]

    for i in barbers_wait:
        cur_time += i
    
    return max(cur_time - 16 * 60, 0)

def five_seats_is_not_enough(outcome):
    pass

def waiting_time_12_00(outcome):
    pass

def monte_carlo(experiment,
                closing_time,
                five_seats_is_not_enough,
                waiting_time_12_00,
                number_of_trials):
    closing_time_s = 0

    for trial in range(number_of_trials):
        state = experiment()
        closing_time_s += closing_time(state)
    
    return closing_time_s / number_of_trials, 0., 0.

monte_carlo(
    experiment=experiment,
    closing_time=closing_time,
    five_seats_is_not_enough=five_seats_is_not_enough,
    waiting_time_12_00=waiting_time_12_00,
    number_of_trials=10000
)
```
